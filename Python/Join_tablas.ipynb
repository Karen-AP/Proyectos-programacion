```python

resumen= foodhub_order.describe(include="all")
res = foodhub_order["cost_of_the_order"].describe(include="all")

minimo = res["min"]
promedio = round(res["mean"],2)
maximo = res["max"]

pd.set_option('display.float_format', '{:.0f}'.format)#para evitar notación exponencial en los resúmenes de pandas
print(f"Resumen de los datos ${resumen}")
pd.set_option('display.float_format', None)# Regresa al formato original de pandas(puede mostrar exponenciales)
print(f"Mínimo del costo de la orden: ${minimo}")
print(f"Máximo del costo de la orden: ${maximo}")
print(f"Promedio del costo de la orden: ${promedio}")
```

```python
#Cuantos valores poseen la caracterìstica "Not given" en la variable "rating"
conteo_valores = foodhub_order["rating"].value_counts()
print(f'Ordenes no calificadas son:{conteo_valores["Not given"]}')

#Cuales restaurantes fueron màs frecuentados de acuerdo con conteo de la variable "restaurant_name"
top_5_restaurantes = foodhub_order["restaurant_name"].value_counts().head(5)
top_5_str = top_5_restaurantes.to_string(header=False)
print(f"Los 5 restaurantes en los que se realizan más pedidos son:\n{top_5_str}")

#tipo de cocina más frecuente
cocina = foodhub_order["cuisine_type"].mode().iloc[0]#mode() Crea una lista con el objeto màs frecuente y iloc[0] toma el primer dato de esa lista y que sea más presentable
print(f"El tipo de cocina más popular es:{cocina}")
```

```python
#Selecciona las ordenes que costaron màs de 20 dolares, calculo del porcentaje, con dos decimales
mayores_a_20 = round((foodhub_order["cost_of_the_order"] > 20).mean() * 100,2)
print(f'Porcentaje de ordenes con un costo mayor a $20: {mayores_a_20}%')

#Promedio de la variable
promedio_orden_f=foodhub_order["delivery_time"].mean()
promedio_orden = round(promedio_orden_f, 2)
print(f'Promedio tiempo de entrega de orden:{promedio_orden} minutos')

#Cuenta las veces que se repite un valor en la variable y muestra los primeros 3
id_cliente = foodhub_order["customer_id"].value_counts().head(3)
id_str = id_cliente.to_string(header=False)#Oculta el encabezado de la variable
print(f"Los 3 clientes que más órdenes realizan son:\n{id_str}")
```

```python
import pandas as pd

# Creación de tres diccionarios con datos ficticios
raw_data_1 = {
    'subject_id' : ['1','2','3','4','5'],
    'first_name' : ['Alex','Amy','Allen','Alice','Ayoung'],
    'last_name' : ['Anderson','Ackerman','Ali','Aoni','Atiches']
}

raw_data_2 = {
    'subject_id' : ['4','5','6','7','8'],
    'first_name' : ['Billy','Brian','Bran','Brice','Betty'],
    'last_name' : ['Bonder','Black','Balwner','Brice','Btisan']
}

raw_data_3 = {
    'subject_id' : ['1','2','3','4','5','7','8','9','10','11'],
    'test_id' : [51,15,15,61,16,14,15,1,61,16]
}

# Creación de DataFrames a partir de los diccionarios
data1 = pd.DataFrame(raw_data_1)
data2 = pd.DataFrame(raw_data_2)
data3 = pd.DataFrame(raw_data_3)

# Concatenación de los DataFrames 'data1' y 'data2' a lo largo de las filas (axis=0)
# Esto combina ambos conjuntos de datos en uno solo sin modificar los índices
all_data = pd.concat([data1, data2])

# Visualización del DataFrame resultante
all_data
```

```python
## Coloca las columnas de 'data2' junto a las de 'data1', sin considerar relaciones entre las filas.
all_data_col = pd.concat([data1, data2], axis=1)
all_data_col
```

```python
# Realizar una fusión (merge) entre 'all_data' y 'data3' usando 'subject_id' como clave
data_join_1 = all_data.merge(data3, on='subject_id')
data_join_1
```

```python
# Fusionar 'all_data' con 'data3' utilizando 'subject_id' como clave, con un INNER JOIN
data_join_2 = all_data.merge(data3, on='subject_id',how='inner')
data_join_2
```

```python
# Fusionar 'all_data' con 'data3' utilizando 'subject_id' como clave, con un OUTER JOIN
data_join_3 = all_data.merge(data3, on='subject_id',how='outer')
data_join_3
```
