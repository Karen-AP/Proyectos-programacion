# Métodos para realizar operaciones con dos o más DF

```python
# Concatenar dataframes
df1 = pd.DataFrame(

    {

        "A": ["A0", "A1", "A2", "A3"],

        "B": ["B0", "B1", "B2", "B3"],

        "C": ["C0", "C1", "C2", "C3"],

        "D": ["D0", "D1", "D2", "D3"],

    },

    index=[0, 1, 2, 3],

)

df2 = pd.DataFrame(

    {

        "A": ["A4", "A5", "A6", "A7"],

        "B": ["B4", "B5", "B6", "B7"],

        "C": ["C4", "C5", "C6", "C7"],

        "D": ["D4", "D5", "D6", "D7"],

    },

    index=[4, 5, 6, 7],

)

df3 = pd.DataFrame(

    {

        "A": ["A8", "A9", "A10", "A11"],

        "B": ["B8", "B9", "B10", "B11"],

        "C": ["C8", "C9", "C10", "C11"],

        "D": ["D8", "D9", "D10", "D11"],

    },

    index=[8, 9, 10, 11]

)
```

```python
print(df1)
print(df2)
print(df3)
```

```python
# Primera forma de concatenar
frames = [df1, df2, df3]
frames
```

```python
# Concatenar con pandas
result = pd.concat([df1, df2, df3])
result
```

```python
# Agregar columnas por indice
df4 = pd.DataFrame(

    {

        "B": ["B2", "B3", "B6", "B7"],

        "D": ["D2", "D3", "D6", "D7"],

        "F": ["F2", "F3", "F6", "F7"],

    },

    index=[2, 3, 6, 7],

)

df4
```

```python
result3 = pd.concat([df1, df4], axis=1) # agrega las columnas y filas sin importar si ya existe en alguno de los dos
result3
```

```python
# Agregar columnas y dejar unicamente registros que coincidan
result4 = pd.concat([df1, df4], axis=1, join="inner")
result4
```

```python
# Agregar columnas y dejar unicamente registros que no coincidan
result5 = pd.concat([df1, df4], axis=1, join="outer")
result5
```

```python
# Agregar filas usando merge
left = pd.DataFrame({"key": [1], "v1": [10]})
left
```

```python
right = pd.DataFrame({"key": [1, 2], "v2": [20, 30]})
right
```

```python
pd.merge(left, right, how="outer", on="key")
```

```python
pd.merge(left, right, how="inner", on="key")
```

# Operaciones agrupadas con Pandas

```python
#### Uso funcion de agrupamiento
df = pd.DataFrame(

    {

        "A": ["foo", "bar", "foo", "bar", "foo", "bar", "foo", "foo"],

        "B": ["one", "one", "two", "three", "two", "two", "one", "three"],

        "C": np.random.randn(8),

        "D": np.random.randn(8),

    }

)

df
```

```python
grouped = df.groupby(["A"]) # Se utiliza únicamente para efectuar operaciones de agrupamiento o resumen
grouped
```

```python
grouped[["C", "D"]].aggregate(np.sum)
```

```python
grouped = df.groupby(["A", "B"])

grouped.aggregate(np.sum)
```

```python
grouped = df.groupby(["A", "B"], as_index=False)

grouped.aggregate(np.sum)
```

```python
# Uso del método transform

#create DataFrame
df = pd.DataFrame({'team': ['A', 'A', 'A', 'A', 'B', 'B', 'B', 'B'],
                   'points': [30, 22, 19, 14, 14, 11, 20, 28]})

#view DataFrame
print(df)
```

```python
# Se agrega una columna nueva con el promedio de puntos por equipo
df['mean_points'] = df.groupby('team')['points'].transform('mean')

print(df)
```

```python
#create new column called percent_of_points
df['percent_of_points'] = df.groupby('team')['points'].transform(lambda x: x/x.sum())

print(df)
```

# Métodos Unstack y Stack



```python
# Se crea un data frame con notas de estudiantes
df = pd.DataFrame({
    'student': ['Alice', 'Bob', 'Carol'],
    'math': [90, 80, 70],
    'science': [85, 95, 90],
    'english': [95, 75, 85]
})

df = df.reset_index(drop=True)

df
```

```python
# Se efectúa el stack al dataframe
series = df.stack() # La función stack convierte un dataframe a una serie usando las etiquetas de las columnas como índices para las filas.

# Se imprime la serie
print(series)
```

```python
# Se aplica unstack a la serie de datos
df = series.unstack() # Lo que hace el método es transformar una serie a un dataframe de pandas

# Print the DataFrame
print(df)
```
